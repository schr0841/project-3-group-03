### TRANSFRORM:
##### Heart Disease dataset

```python
# Make a copy of the heart disease indicators dataset
hd_df = heart_disease_2022_df.copy()
hd_df.head()
```

```python
# Split the 'RaceEthnicityCategory' column on the comma into two columns
hd_df[['Race', 'Ethnicity']] = hd_df['RaceEthnicityCategory'].str.split(',', expand=True)
hd_df.head()
```

```python
# Drop the 'RaceEthnicityCategory' column
hd_df = hd_df.drop(columns=('RaceEthnicityCategory'))
hd_df.head()
```

```python
# Check the uniques vales in the State column
states = hd_df['State'].unique()
states
```

```python
# Dropping rows (to keep all states and the District of Columbia ONLY)
hd_df = hd_df[~hd_df['State'].isin(['Guam', 'Puerto Rico', 'Virgin Islands'])] #.reset_index(drop=True)
hd_df.head()
```

```python
# Confirm the regions were dropped
states = hd_df['State'].unique()
states
```

```python
# Replacing the Yes/No values to 0/1
columns_to_replace = [
    'HadHeartAttack', 'HadAngina', 'HadStroke', 'HadAsthma', 'HadSkinCancer',
    'HadCOPD', 'HadDepressiveDisorder', 'HadKidneyDisease', 'HadArthritis',
    'HadDiabetes', 'DeafOrHardOfHearing', 'BlindOrVisionDifficulty',
    'DifficultyConcentrating', 'DifficultyWalking', 'DifficultyDressingBathing',
    'DifficultyErrands', 'AlcoholDrinkers', 'ChestScan', 'HIVTesting',
    'FluVaxLast12', 'PneumoVaxEver', 'HighRiskLastYear', 'CovidPos', 'PhysicalActivities'
]

# Replace values in specified columns
for column in columns_to_replace:
    hd_df[column] = hd_df[column].replace({'No': 0, 'Yes': 1})

hd_df.head()
```

```python
# Check the unique values of HadDiabetes column
hadDiabetes_values = hd_df['HadDiabetes'].unique()
hadDiabetes_values
```

```python
# Check the unique values of CovidPos column
covidPos_values = hd_df['CovidPos'].unique()
covidPos_values
```

```python
# Replacing the other values of the HadDiabetes and CovidPos columns to numeric values
hd_df['HadDiabetes'] = hd_df['HadDiabetes'].replace({'Yes, but only during pregnancy (female)': 3,
       'No, pre-diabetes or borderline diabetes': 4})
hd_df['CovidPos'] = hd_df['CovidPos'].replace({'Tested positive using home test without a health professional': 3})
```

```python
convert_list = [
    'PhysicalHealthDays', 'MentalHealthDays', 'SleepHours', 'PhysicalActivities',
    'HadHeartAttack', 'HadAngina', 'HadStroke', 'HadAsthma', 'HadSkinCancer', 
    'HadCOPD', 'HadDepressiveDisorder', 'HadKidneyDisease', 'HadArthritis', 'HadDiabetes',
    'DeafOrHardOfHearing', 'BlindOrVisionDifficulty', 'DifficultyConcentrating', 'DifficultyWalking', 
    'DifficultyDressingBathing', 'DifficultyErrands', 'AlcoholDrinkers', 'ChestScan', 'HIVTesting',
    'FluVaxLast12', 'PneumoVaxEver', 'HighRiskLastYear', 'CovidPos'
]

# Convert specified columns to integers
for column in convert_list:
    hd_df[column] = hd_df[column].astype(int)

# Check the data types of the specified columns
hd_df[convert_list].dtypes
```

##### Transforming the Census Data Population Estimates (Age 18+)

```python
# Set number of rows to be displayed
pd.set_option('display.max_rows', 68)

# Make a copy of the population estimates dataset
census_df = census_data_2022_df.copy()
census_df
```

```python
# Rename the census_df columns
census_df.rename(columns = {
    0: 'State', 
    1: 'TotalPop', 
    2: 'TotalNum', 
    3: 'TotalPercent', 
}, inplace=True)

census_df.head(10)
```

```python
# Drop Nan values from the population estimates (contain descriptions of the dataset)
census_df.dropna(inplace=True)
census_df.head(10)
```

```python
# Dropping rows (to keep all states and the District of Columbia ONLY)
us_states_df = census_df[~census_df['State'].isin(['United States', 'Northeast', 'Midwest', 'South', 'West', 'Puerto Rico'])].reset_index(drop=True)
us_states_df.head()
```

```python
us_states_df['State'] = us_states_df['State'].replace(r'^\.', '', regex=True)
us_states_df.head(10)
```

```python
# Check the dataypes
us_states_df.dtypes
```

```python
# Convert the Total Population and Number columns to 'integer' and Total Percentage column to a `float` data type.
us_states_df = us_states_df.astype({'TotalPop': 'int', 'TotalNum': 'int', 'TotalPercent': 'float'})
us_states_df.head()
```

```python
# Confirm the dataypes
us_states_df.dtypes
```

```python
# Define regions
us_regions = {
    'Northeast': ['Connecticut', 'Maine', 'Massachusetts', 'New Hampshire', 'Rhode Island', 'Vermont', 'New Jersey', 'New York', 'Pennsylvania'],
    'Midwest': ['Illinois', 'Indiana', 'Michigan', 'Ohio', 'Wisconsin', 'Iowa', 'Kansas', 'Minnesota', 'Missouri', 'Nebraska', 'North Dakota', 'South Dakota'],
    'South': ['Delaware', 'District of Columbia', 'Florida', 'Georgia', 'Maryland', 'North Carolina', 'South Carolina', 'Virginia', 'West Virginia', 'Alabama', 'Kentucky', 'Mississippi', 'Tennessee', 'Arkansas', 'Louisiana', 'Oklahoma', 'Texas'],
    'West': ['Arizona', 'Colorado', 'Idaho', 'Montana', 'Nevada', 'New Mexico', 'Utah', 'Wyoming', 'Alaska', 'California', 'Hawaii', 'Oregon', 'Washington']
}

# Create a new 'Region' column based on the states
def assign_region(state):
    if state in us_regions['Northeast']:
        return 'Northeast'
    elif state in us_regions['Midwest']:
        return 'Midwest'
    elif state in us_regions['South']:
        return 'South'
    elif state in us_regions['West']:
        return 'West'
    else:
        return 'Other'

# Assuming usa_states_df is your DataFrame containing the USA states
census_df = pd.DataFrame(us_states_df)
census_df['Region'] = census_df['State'].apply(assign_region)

# Create a new DataFrame for the 'Regions' table
us_regions_df = census_df[['Region', 'State', 'TotalPop', 'TotalNum','TotalPercent']]
us_regions_df
```

```python
# Export hd_df and us_regions_df as CSV files
hd_df.to_csv("Resource/hd_df.csv", index=False)
us_regions_df.to_csv("Resource/us_regions_census.csv", index=False)
```
