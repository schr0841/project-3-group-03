# Answer Research Questions 

### Is this survey data uniformly distributed according to state population?

```python
with engine.connect() as conn:
    q3 = conn.execute(text('''select hd.state, hd.count survey_size, u.totalnum as numover18, 
    (cast(hd.count as numeric)/cast(u.totalnum as numeric)) as samplepct from 
(SELECT state, COUNT(*) from hd_df GROUP BY state ORDER BY COUNT(*) DESC) as hd 
INNER JOIN us_regions_census u ON hd.state=u.state ORDER BY samplepct DESC''')).fetchall()

q3
```

### Research Question 1: How does the incidence of heart disease vary across the different states/regions in the US?


### State Analysis

```python
with engine.connect() as conn:
    q4 = conn.execute(text('''select state, cast(sum(hadheartattack) as numeric)/cast(count(hadheartattack) as numeric) as heartattack_pct
from hd_df 
group by state
order by heartattack_pct DESC''')).fetchall()

q4
```

### Region Analysis 

```python
with engine.connect() as conn:
    q5 = conn.execute(text('''select u.region, cast(sum(h.hadheartattack) as numeric)/cast(count(h.hadheartattack) as numeric) as heartattack_pct from hd_df h 
INNER JOIN us_regions_census u 
on h.state=u.state
GROUP BY u.region
ORDER BY heartattack_pct DESC''')).fetchall()

q5
```

### Question 3: Do males or females have a higher incidence of heart disease?

```python
with engine.connect() as conn:
    q6 = conn.execute(text('''select sex, sum(hadheartattack), 
round(cast(sum(hadheartattack) as numeric)/cast(count(hadheartattack) as numeric), 2) as heartdiseasepct 
from hd_df 
group by sex
order by heartdiseasepct DESC''')).fetchall()

q6
```
